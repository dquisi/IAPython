```python
import json
from ibm_watson import LanguageTranslatorV3
from ibm_cloud_sdk_core.authenticators import IAMAuthenticator
```

## Traducci√≥n dle lenguaje

```python
authenticator = IAMAuthenticator('bYAflzALvlFROWuZwN5mb5SpZEtBYM8aTxg-2P1MhMR-')
language_translator = LanguageTranslatorV3(
    version='2018-05-01',
    authenticator=authenticator)
language_translator.set_service_url('https://api.us-south.language-translator.watson.cloud.ibm.com/instances/8390ca50-4099-43e2-910e-c1f5e7cd41a5')

## Translate
translation = language_translator.translate(
    text='Hola mundo desde la UPS', model_id='es-en').get_result()
print(json.dumps(translation, indent=2, ensure_ascii=False))
```

## Identificaciones del lenguaje

```python
 language = language_translator.identify(
     'Hola mundo desde la UPS').get_result()
 print(json.dumps(language, indent=2))
```
